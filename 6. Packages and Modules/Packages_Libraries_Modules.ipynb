# Basic Definitions

* module: any *.py file. Its name is the file name.
* built-in module: a “module” (written in C) that is compiled into the Python interpreter, and therefore does not have a *.py file.
* package: any folder containing a file named ````__init__.py```` in it. Its name is the name of the folder.
* in Python 3.3 and above, any folder (even without a ````__init__.py```` file) is considered a package
* object: in Python, almost everything is an object - functions, classes, variables, etc.

#What happens when you import a module or a package?

* When a module is imported, Python runs all of the code in the module file. 
* When a package is imported, Python runs all of the code in the package’s __init__.py file, if such a file exists. 
* All of the objects defined in the module or the package’s __init__.py file are made available to the importer.

## There are 4 different syntaxes for writing import statements.

* import ````<package>````
* import ````<module>````
* from ````<package>```` import ````<module or subpackage or object>````
* from ````<module>```` import ````<object>````

# Importing packages

Let's take a look at BeautifulSoup, a web scraping package for python. Beautiful soup has some dependencies that we're going to need to import as well, such as ````requests```` to handle the url.

I am assuming you have some experience with web scraping. If not, you can learn more about html structure [here](https://www.w3schools.com/html/html_intro.asp)

`````python
from bs4 import BeautifulSoup # By importing BeautifulSoup specifically, we don't have to use bs4.BeautifulSoup 
import requests

import matplotlib.pyplot as plt
import numpy as np
`````

`````python
webpage = requests.get("https://content.codecademy.com/courses/beautifulsoup/cacao/index.html")
soup = BeautifulSoup(webpage.content, "html.parser")
`````

Let's use the find_all function from BS to create a list of all REF numbers in the sample website:

`````python
example_column = soup.find_all(attrs={"class": "REF"})
example_list = []
for x in example_column[1:] : 
   example_list.append(float(x.get_text()))
`````

`````python
print(example_list)
`````

# Try it out

Create two lists: One with the Ratings called ````ratings```` and one with the percentage of cocoa called ````cocoa_percentage````.

Hint: You can play around with this by looking at the page source on codeacademy. 

`````python
percent = soup.find_all(attrs={"class": "CocoaPercent"})
percentage_list = []
for x in percent[1:] : 
   percentage_list.append(float(x.get_text()[0:2]))
`````

`````python
ratings = soup.find_all(attrs={"class": "Rating"})
ratings_list = []
for x in ratings[1:] : 
   ratings_list.append(float(x.get_text()))
`````

`````python
print(percentage_list)
print(ratings_list)
`````

# Now let's Play around with this data.

We are going to create a dataframe with two columns: Ratings and CocoaPercentage.

We will learn a lot more about dataframes in a couple of weeks but for now, let's just import pandas as pd:

`````python
import pandas as pd
`````

And let's create a dictionary from the two lists we created above:

`````python
data = {"Rating": ratings_list, "CocoaPercentage": percentage_list}
`````

Now we can create a quick dataframe using pandas:

`````python
df = pd.DataFrame.from_dict(data)
`````

`````python
df.sample(5)
`````

Now we're going to use numpy to analyse our data. 

We are going to learn a lot more about numpy in a couple of weeks, but for now let's just import it as np and use a function called polyfit:

`````python
import numpy as np
`````

`````python
z = np.polyfit(df.CocoaPercentage, df.Rating, 1)
`````

Now we will use numpy to define a polynomial function for this data:

`````python
line_function = np.poly1d(z)
`````

Finally, let's use pyplot to visualise this data:

Pyplot is part of matplotlib. We are going to learn a lot more about matplotlib in a couple of weeks, but for now let's just import it as plt:

`````python
from matplotlib import pyplot as plt
`````

Another way to import pyplot would be to use dot notation:

`````python
import matplotlib.pyplot as plt
`````

Now let's use a scatterplot:

`````python
plt.scatter(df.CocoaPercentage, df.Rating)
plt.title('Cocoa Percentage & Ratings Correlation')
plt.xlabel('Cocoa Percentage (%)')
plt.ylabel('Ratings')
plt.plot(df.CocoaPercentage, line_function(df.CocoaPercentage), "r--")
plt.show()
`````
